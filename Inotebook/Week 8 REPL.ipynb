# Properties

```python
class Ship:
    
    def __init__(self):
        self.angle = 90
        self.alive = True
        self.fuel = 100
        self.speed = 10
        
    def rotate(self, delta):
        self.angle += delta

    def display(self):
        print("Speed: {} Angle: {} Alive: {} Fuel: {}" 
              .format(self.speed, self.angle, self.alive, self.fuel))

    """
    def get_fuel(self):
        return self.fuel

    def set_fuel(self, fuel):
        if fuel < 0:
            self.fuel = 0
        else:
            self.fuel = fuel
    """
    
    @property
    def fuel(self):
        return self._fuel

    @fuel.setter
    def fuel(self, fuel):
        if fuel < 0:
            self._fuel = 0
        else:    
            self._fuel = fuel

    @property
    def speed(self):
        return self._speed

    @speed.setter
    def speed(self, speed):   
        if speed < 0:
            self._speed = 0
        elif speed > 20:
            self._speed = 20
        else:
            self._speed = speed

    @property
    def angle(self):
        return self._angle
        
    @angle.setter
    def angle(self, angle):
        self._angle = angle % 360

    @property
    def can_move(self):
        if self.alive and self.fuel > 0:
            return True
        else:
            return False
            
def main():
    ship = Ship()
    #ship.setSpeed(10)
    #ship.speed = -5
    ship.display()

    ship.speed = 50
    ship.angle = 370
    ship.fuel = 1000
    
    ship.display()
    
    ship.rotate(-20)
    ship.display()
    
    ship.fuel = -1
    ship.display()
    print("Can move: {}" .format(ship.can_move))

    ship.fuel = 100
    ship.display()
    print("Can move: {}" .format(ship.can_move))
    
    #ship.can_move = True
    


if __name__ == "__main__":
    main()
```

# BInary Search Tree

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
class BinarySearchTree:
    def __init__(self):
        self.root = Node(None)
        
    def insert(self, data):
        # Special Case - Empty Tree
        if self.root.data == None:
            self.root.data = data
        # Search from Root
        else:
            self._insert(data, self.root)
    
    def _insert(self, data, node):
        # If the data we are adding is less then the current node:
        # a) If there is nothing to the left, then put the new node there
        # b) Otherwise, follow it
        #
        # Otherwise:
        # a) If there is nothing to the right, then put the new node there
        # b) Otherwise, follow it
        if data < node.data:
            # Found empty spot
            if node.left == None:
                node.left = Node(data)
            # Follow left link
            else:
                self._insert(data, node.left)
        # Look right
        else:
            # Found empty spot
            if node.right == None:
                node.right = Node(data)
            # Follow right link
            else:
                self._insert(data, node.right)
    
    def display_forward(self):
        # Special Case - Empty Tree
        if self.root.data == None:
            print("Empty")
        # Search from the root
        else:
            self._display_forward(self.root)
            
    def _display_forward(self, node):
        # If the node exists:
        # a) Follow the tree to the left (prints all the smaller numbers)
        # b) Print out the current node
        # c) Follow the tree to the right (prints all the larger numbers)
        if node != None:
            # Go all the way to the left (smaller numbers)
            self._display_forward(node.left)
            # Print out the number
            print(node.data)
            # Go all the way to the right (larger numbers)
            self._display_forward(node.right)
            
    def display_backward(self):
        # Special Case - Empty Tree
        if self.root.data == None:
            print("Empty")
        # Search from the root
        else:
            self._display_backward(self.root)
            
    def _display_backward(self, node):
        # If the node exists:
        # a) Follow the tree to the right (prints all the larger numbers)
        # b) Print out the current node
        # c) Follow the tree to the right (prints all the smaller numbers)        
        if node != None:
            # Go all the way to the right (larger numbers)
            self._display_backward(node.right)
            # Print out the number
            print(node.data)
            # Go all the way to the left (smaller numbers)
            self._display_backward(node.left)
            
def main():
    bst = BinarySearchTree()
    choice = ""
    while choice != "4":
        print("1) Insert")
        print("2) Display Forward")
        print("3) Display Backward")
        print("4) Quit")
        choice = input("> ")
        if choice == "1":
            data = int(input("New value: "))
            bst.insert(data)
        elif choice == "2":
            bst.display_forward()
        elif choice == "3":
            bst.display_backward()
        print()

if __name__ == "__main__":
    main()
```
